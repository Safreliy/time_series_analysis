## Лабораторная работа №1

Импортируем библиотеки

```python
import pandas as pd
import plotly.express as px
import numpy as np
```

Создаём Pandas DataFrame из файла IBM.csv

```python
file_path = 'IBM.csv'
df = pd.read_csv(file_path)
```

Считаем статистические характеристики

```python
min_price = df['Close'].min()
max_price = df['Close'].max()
mean_price = df['Close'].mean()
median_price = df['Close'].median()
moda_price = df['Close'].value_counts().idxmax()
std_deviation = df['Close'].std()

print(f"Минимальная стоимость: {min_price}")
print(f"Максимальная стоимость: {max_price}")
print(f"Средняя стоимость: {mean_price}")
print(f"Медианная стоимость: {median_price}")
print(f"Мода: {moda_price}")
print(f"Стандартное отклонение: {std_deviation}")
```

Отрисуем график стоимости акций

```python
fig = px.line(df, x='Date', y='Close', title='Стоимость акций IBM')
fig.update_xaxes(title='Дата')
fig.update_yaxes(title='Стоимость')
fig.show()
```

Отрисуем график гистограммы

```python
fig = px.histogram(df, x='Close', nbins=int(1+np.floor(np.log2(len(df)))), title='Гистограмма акций IBM')
fig.update_xaxes(title='Стоимость')
fig.update_yaxes(title='Количество')
fig.show()
```

Отрисуем график boxplot

```python
fig = px.box(df, y='Close', title='Боксплот акций IBM')
fig.update_yaxes(title='Стоимость')
fig.show()
```
